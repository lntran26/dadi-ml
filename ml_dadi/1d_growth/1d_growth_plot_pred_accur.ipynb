```python
# Load files for ML models trained on new_func data
# load testing data set
list_test_dict = pickle.load(open('data/new_func/test_data','rb'))
# load list of train rfr
list_rfr = pickle.load(open('data/new_func/list_rfr','rb'))
# load list of trained mlpr with lbfgs solver
list_mlpr_lbfgs = pickle.load(open('data/new_func/list_mlpr_lbfgs','rb'))
# load list of trained mlpr with adam solver
list_mlpr_adam = pickle.load(open('data/new_func/list_mlpr_adam','rb'))
```

```python
# Plot MLPR 2x2 training & testing var, with colorbar
list_test_dict_clipped = [list_test_dict[0], list_test_dict[2]]
list_mlpr_clipped = [list_mlpr_lbfgs[0], list_mlpr_lbfgs[2]]
params = ['ν', 'T']
model_name = 'MLPR lbfgs'
logs = [True, False]
# (50, 20), (30, 80), (15,4,40), (15,13)
# (50, 80),                     (18,15)
size = ((50, 80), (30, 80), (15,4,40), (16.5,13))
# size_format = ( (title_font_size, title_pad), (axis_font_size, axis_pad),
#                   (single_size_tuple), fig_size)
# single_size_tuple = (dots_size, line_width, font_size)
plot_accuracy_multi(list_mlpr_clipped, list_test_dict_clipped, params, model_name, logs, size, r2=False, c=True)
# save fig
for i in range(len(params)):
    plt.figure(i+1).savefig(f'results/MLPR/new_func/{params[i]}_2x2_colors.png')
```
